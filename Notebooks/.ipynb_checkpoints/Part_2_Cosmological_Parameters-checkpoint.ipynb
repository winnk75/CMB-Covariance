### Introduction to Cosmological Parameters

The CMB power spectrum is generated from a variety of cosmological parameters, some more important than others. There are generally six "primary" parameters, and dozens more (depending on the specific model) that determine the finer points of the spectrum. New Higgs Inflation, for example, uses several parameters to mathematically define the nature of the inflation. We'll worry about the specifics of New Higgs in the next notebook. First, we're going to play around with some more standard parameters.

There are already a great deal of simulated power spectra included in the data directories, so let's just load up a bunch of data and plot the power spectra for various parameters.

### Varying $H_0$
Varying $H_0$ can do a lot of funky things to the power spectrum. Generally, it moves the first peak back and fourth (the first peak has to do with the cosmic horizon at inflation), as well as the higher modes. By tweaking $H_0$ and some of the mass fractions, you can slide the peaks around without changing the amplitude.

```python
plt.plot(axis, data_h0_0_40, label='$H_0$ = 45.0', color='#E56100', linestyle='dashdot')
plt.plot(axis, data_h0_0_50, label='$H_0$ = 50.0', color='#AB642F', linestyle='dashed')
plt.plot(axis, data_nominal, label='$H_0$ = 67.3 *', color='#72675F')
plt.plot(axis, data_h0_0_80, label='$H_0$ = 80.0', color='#396A8F', linestyle='dashed')
plt.plot(axis, data_h0_0_90, label='$H_0$ = 90.0', color='#006EBF', linestyle='dashdot')


plt.legend(loc='upper right')
plt.xlabel('$\ell$')
plt.ylabel(r'$\frac {\ell (\ell+1)} {2 \pi}$ $C_\ell$  [$\mu$K$^2$]')
plt.title('Simulated Power Spectra at Various $H_0$ Values')
plt.show()

#plt.savefig('./Wide-Varied-h_0-TT_dash.pdf', bbox_inches='tight')
#plt.savefig('./Wide-Varied-h_0-TT_dash.png', bbox_inches='tight', dpi=300)
plt.clf()  # clears the plot memory so we don't overlap into other plots
```

### Varying $\tau$
The optical depth of recombination more or less just changes the amount of light (and thus amplitude) across all modes.

```python
# Now some plots for tau...

plt.plot(axis, data_tau_0_06, label=r'$\tau$ = 0.060', color='#E56100', linestyle='dashdot')
plt.plot(axis, data_tau_0_07, label=r'$\tau$ = 0.070', color='#AB642F', linestyle='dashed')
plt.plot(axis, data_nominal, label=r'$\tau$ = 0.078 *', color='#72675F')
plt.plot(axis, data_tau_0_09, label=r'$\tau$ = 0.090', color='#396A8F', linestyle='dashed')
plt.plot(axis, data_tau_0_10, label=r'$\tau$ = 0.100', color='#006EBF', linestyle='dashdot')


plt.legend(loc='upper right')
plt.xlabel('$\ell$')
plt.ylabel(r'$\frac {\ell (\ell+1)} {2 \pi}$ $C_\ell$  [$\mu$K$^2$]')
plt.title(r'Simulated Power Spectra at Various $\tau$ Values')
plt.show()

#plt.savefig('./Varied-tau-TT.pdf', bbox_inches='tight')
#plt.savefig('./Varied-tau-TT.png', bbox_inches='tight', dpi=300)
plt.clf()
```

### Varying $n_s$
The scalar spectral index has to do with the speed of inflation. If the index were flat, inflation would be very quick (?) as the quantum fluctuations are the same on all sizes. Slower inflation means that quantum fluctuations are not as pronounced later in inflation.

It is helpful to plot the index *without* the peaks, to see the slope of the line formed by the index, without the specific anisotrpies.

```python
# n_s plots

plt.plot(axis, data_ns_0_80, label=r'$n_s$ = 0.80', color='#E56100', linestyle='dashdot')
plt.plot(axis, data_ns_0_90, label=r'$n_s$ = 0.90', color='#AB642F', linestyle='dashed')
plt.plot(axis, data_nominal, label=r'$n_s$ = 0.9658 *', color='#72675F')
plt.plot(axis, data_ns_1_00, label=r'$n_s$ = 1.00', color='#396A8F', linestyle='dashed')
plt.plot(axis, data_ns_1_10, label=r'$n_s$ = 1.10', color='#006EBF', linestyle='dashdot')


plt.legend(loc='upper right')
plt.xlabel('$\ell$')
plt.ylabel(r'$\frac {\ell (\ell+1)} {2 \pi}$ $C_\ell$  [$\mu$K$^2$]')
plt.title(r'Simulated Power Spectra at Various $n_s$ Values')
plt.show()

#plt.savefig('./Varied-n_s-TT.pdf', bbox_inches='tight')
#plt.savefig('./Varied-n_s-TT.png', bbox_inches='tight', dpi=300)
plt.clf()
```

```python
# linear n_s plots, clipped the high L to avoid divide by zero issues

plt.loglog(axis[0:2000], data_ns_0_80[0:2000]/data_nominal[0:2000], label=r'$n_s$ = 0.80', color='#E56100')
plt.loglog(axis[0:2000], data_ns_0_90[0:2000]/data_nominal[0:2000], label=r'$n_s$ = 0.90', color='#AB642F')
plt.loglog(axis[0:2000], data_nominal[0:2000]/data_nominal[0:2000], label=r'$n_s$ = 0.9658 *', color='#72675F')
plt.loglog(axis[0:2000], data_ns_1_00[0:2000]/data_nominal[0:2000], label=r'$n_s$ = 1.00', color='#396A8F')
plt.loglog(axis[0:2000], data_ns_1_10[0:2000]/data_nominal[0:2000], label=r'$n_s$ = 1.10', color='#006EBF')

plt.legend(loc='upper right')
plt.xlabel('$\ell$')
plt.title(r'Flattened $n_s$ Spectral Index [Log]')
plt.show()

#plt.savefig('./Varied-n_s-flattened-TT.pdf', bbox_inches='tight')
#plt.savefig('./Varied-n_s-flattened-TT.png', bbox_inches='tight', dpi=300)
plt.clf()
```

### Varying $\Omega_{\Lambda}$, $\Omega_b$

```python
# Omega_lambda plots

plt.plot(axis, data_omega_lambda_0_50, label=r'$\Omega_{\Lambda}$ = 0.50', color='#E56100', linestyle='dashdot')
plt.plot(axis, data_omega_lambda_0_60, label=r'$\Omega_{\Lambda}$ = 0.60', color='#AB642F', linestyle='dashed')
plt.plot(axis, data_nominal, label=r'$\Omega_{\Lambda}$ = 0.69 *', color='#72675F')
plt.plot(axis, data_omega_lambda_0_80, label=r'$\Omega_{\Lambda}$ = 0.80', color='#396A8F', linestyle='dashed')
plt.plot(axis, data_omega_lambda_0_90, label=r'$\Omega_{\Lambda}$ = 0.90', color='#006EBF', linestyle='dashdot')

plt.legend(loc='upper right')
plt.xlabel('$\ell$')
plt.ylabel(r'$\frac {\ell (\ell+1)} {2 \pi}$ $C_\ell$  [$\mu$K$^2$]')
plt.title(r'Simulated Power Spectra at Various $\Omega_{\Lambda}$ Values ($\Omega_b / \Omega_c$ fixed)')
plt.show()

#plt.savefig('./Varied-omega_lambda-TT.pdf', bbox_inches='tight')
#plt.savefig('./Varied-omega_lambda-TT.png', bbox_inches='tight', dpi=300)
plt.clf()

# Inverse degeneracy with H_0!!
```
