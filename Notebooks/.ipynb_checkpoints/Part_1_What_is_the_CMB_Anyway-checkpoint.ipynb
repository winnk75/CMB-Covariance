### What is New Higgs Inflation and why you should care
New Higgs Inflation is a class of inflation models that places the Higgs Boson from particle physics at the center of astrophysics and cosmology. Effectively, this model attempts to bridge the gap between particle physics and astrophysics by asking whether or not the Higgs Boson, the capstone of Standard Model particle physics which gives everything in the universe mass, can also cause the period of rapid inflation immediately after the Big Bang, which ultimately created the conditions for the universe as we know it to form. What causes cosmic inflation and why it happened are some of the largest unsolved questions in astrophysics -- it's literally Nobel Prize level stuff.

In one sentence: We're trying to unite the smallest of the small (particle physics) with the largest of the large (life, the universe, and everything) through a single particle. So it's pretty exciting, even to someone who knows nothing about physics.

## Introduction to the CMB

We're going to start with a very short history of the early universe, for some context.

<br>

$0$ sec
- The Big Bang singularity. Not much more to say here, we don't really know.

~$10^{-35}$ sec
- Cosmic Inflation begins, expanding the universe in size (and, consequently, energy) by approxomately $e^{60}$ times. This size difference is comperable to the difference between a virus and a galaxy.
- Small fluctuations in energy (or mass, $e = mc^2$) due to quantum mechanical uncertainty (from Heisenberg: $\frac{\hbar}{2} \leq \Delta x \Delta p$) are expanded across the suddenly giagantic universe.
- Causal disconnect occurs. The universe expands ''faster than the speed of light.'' That's in quotes because nothing broke the speed of light. The spacetime of the universe can expand as quickly or slowly as it wants, and the only change is that the *distances* between objects gets larger. So we have the illusion of things moving faster than the speed of light. Two particles in thermal equillibrium one milimeter apart are now a few kiloparsecs away from each other, in just a fraction of a second. The point is that these particles could no longer possibly be in thermal equillibrium, they're too far apart now. This is what we mean by causal disconnect: the particles are now a few million lightyears apart, so they can't ''talk'' to each other anymore.

~$10^{-33}$ sec
- Quantum fluctuations are ''frozen-in'' by the causal disconnect. The universe went from being very small to very big in almost no time at all, and the energy fluctuations due to the Heisenberg Uncertainty principle don't have time to smooth out (Heisenberg's can also be written as $\frac{\hbar}{2} \leq \Delta E \Delta t$). So we have regions of the universe with a little bit more energy/mass, and regions with a little bit less.

We now have a large, hot, ionized universe. As things start to cool, we get bigger and bigger particles. At about a second, everything is a soup of quarks and other particles. After a minute or two (literally one or two minutes) we have protons and neutrons, which start to combine. After about ten minutes, the unbound netrons decay, giving off more energy. The ones that were lucky enough to find protons to bind to form the first helium in the universe. Most protons, however, did not bind, so there's a great deal more ionized hydrogen (just a proton) than ionized helium.

Some very interesting stuff happens here, but for our purposes we can fast forward until ~$380,000$ years. It's important to note that during the hot, ionized early universe, unbound electrons are everywhere. And photons undergo Thompson scattering very easily with unbound electrons. There is light in the universe, but it's incoherent, bouncing from one electron to another. Everything is a big, hot, opaque cloud. But the universe has been expanding and cooling for a few hundred thousand years, and eventually it is cool enough that the electrons can bind to protons, forming neutral hydrogen gas. Hydrogen gas only absorbs and emits along a few select frequencies consistent with the electron energy levels, so where there once was a brick wall of free electrons, there is now a glass window of neutral hydrogen. Photons can suddenly free stream coherently. This is the Cosmic Microwave Background (CMB).

<img src="Images/cmb_wmap.png" alt="The cmb!" title="CMB" />

The CMB is the first visable light in the universe, and tells us all kinds of things about the universe immediately after the big bang. See how it's not perfectly smooth? Those hot spots and cold spots? Those are the quantum fluctuations we talked about. It's Hiesenberg's Uncertainty Principle, etched into a surface with an area many hundreds of billions of lightyears. This is perhaps the largest clue we have that cosmic inflation actually occured.

## CMB Analysis with Spherical Harmonics

The CMB has a temperature of about $2.7 \ \textrm{K}$, and the range of fluctuations shown in the map above are roughly $\pm 200 \ \mu \textrm{K}$. While this map is what is observed by our telescopes, it is of lesser use to cosmologists (and basically everyone), who prefer to study the CMB in fourier space.

We convert this map into a power spectrum, which is effectively a 2d Fourier transform. There are a few different ways to perform a fourier transform, so let's consider what might be the most useful. The CMB is on the surface of a sphere, and best way to encode wave information on the surface of a sphere is to use spherical harmonics. 

Below, an image of some orthonormal spherical harmonics from __[here](http://www.helsinki.fi/~hkurkisu/cosmology/Cosmo12.pdf)__.

<img src="Images/spherical_harmonics.png" alt="The cmb!" title="CMB" style="width: 600px" />

Since the set of spherical harmonics forms a **complete basis**, we can express *any function* that exists on the surface of a sphere as a vector or different spherical harmonics. Don't believe me? Here's a topographical map of the Earth built using spherical harmonics (__[source](https://www.researchgate.net/figure/Fig-S6-Spherical-harmonic-models-for-the-topography-of-Titan-Earth-and-Mars-For_fig8_317079579)__). The more you add, the better the map is.

## The CMB Power Spectrum

Let us now introduce the CMB power spectrum. The horizontal axis is the spherical harmonic, $\ell$. So farther right is larger $\ell$ and smaller angular scale. The vertical axis is just the amplitude. So effectively it's how present is this spherical harmonic or frequency in the data? The canonical CMB plot that everyone uses is shown below.

### Defining functions to generate a CMB sky

I think it's a worthwhile exercise to build a simulated patch of CMB sky from a source power spectrum. You'll never do this for the analysis, but sometimes you get lost in the power spectrum and forget what it actually represents. So below, I've taken some functions from Renee's summer school notebooks, which will take our power spectra, and perform an inverse Fourier transform, with some random phase added.



```python
## variables to set up the size of the map
N = 2**10.  # this is the number of pixels in a linear dimension
            ## since we are using lots of FFTs this should be a factor of 2^N
pix_size  = 0.5 # size of a pixel in arcminutes

## variables to set up the map plots
c_min = -400  # minimum for color bar
c_max = 400   # maximum for color bar
X_width = N*pix_size/60.  # horizontal map width in degrees
Y_width = N*pix_size/60.  # vertical map width in degrees


def make_CMB_T_map(N,pix_size,ell,DlTT):
    "makes a realization of a simulated CMB sky map"

    # convert Dl to Cl
    ClTT = DlTT * 2 * np.pi / (ell*(ell+1.))
    ClTT[0] = 0.
    ClTT[1] = 0.

    # make a 2d coordinate system
    ones = np.ones(int(N))
    inds  = (np.arange(N)+.5 - N/2.) /(N-1.)
    X = np.outer(ones,inds)
    Y = np.transpose(X)
    R = np.sqrt(X**2. + Y**2.)
    
    # now make a 2d CMB power spectrum
    ell_scale_factor = 2. * np.pi / (pix_size/60. * np.pi/180.)
    ell2d = R * ell_scale_factor
    ClTT_expanded = np.zeros( int(ell2d.max()+1) )
    ClTT_expanded[0:(ClTT.size)] = ClTT
    CLTT2d = ClTT_expanded[ell2d.astype(int)]
    print CLTT2d
    print CLTT2d.shape
    print CLTT2d.max()
    print np.mean(CLTT2d)
    ## make a plot of the 2d cmb power spectrum, note the x and y axis labels need to be fixed
    #Plot_CMB_Map(CLTT2d**2. *ell2d * (ell2d+1)/2/np.pi,0,np.max(CLTT2d**2. *ell2d * (ell2d+1)/2/np.pi)/10.,ell2d.max(),ell2d.max())  ###
 
    # now make a realization of the CMB with the given power spectrum in fourier space
    ramdomn_array_for_T = np.fft.fft2(np.random.normal(0,1,(int(N),int(N))))    
    FT_2d = np.sqrt(CLTT2d) * ramdomn_array_for_T
    ## make a plot of the 2d cmb simulated map in fourier space, note the x and y axis labels need to be fixed
    #Plot_CMB_Map(np.real(np.conj(FT_2d)*FT_2d*ell2d * (ell2d+1)/2/np.pi),0,np.max(np.conj(FT_2d)*FT_2d*ell2d * (ell2d+1)/2/np.pi),ell2d.max(),ell2d.max())  ###
    CMB_T = np.fft.ifft2(np.fft.fftshift(FT_2d)) /(pix_size /60.* np.pi/180.)
    CMB_T = np.real(CMB_T)

    ## return the map
    return(CMB_T)
  ###############################

def Plot_CMB_Map(Map_to_Plot,c_min,c_max,X_width,Y_width):
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    print("map mean:",np.mean(Map_to_Plot),"map rms:",np.std(Map_to_Plot))
    plt.figure(figsize=(8,8))
    im = plt.imshow(Map_to_Plot, interpolation='bilinear', origin='lower',cmap=cm.RdBu_r)
    im.set_clim(c_min,c_max)
    ax=plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)

    cbar = plt.colorbar(im, cax=cax)
    #cbar = plt.colorbar()
    im.set_extent([0,X_width,0,Y_width])
    plt.ylabel('angle $[^\circ]$')
    plt.xlabel('angle $[^\circ]$')
    cbar.set_label('tempearture [uK]', rotation=270)
    #plt.savefig('foo.png', bbox_inches='tight')
    plt.show()
    return(0)
  ###############################


```

Above, we define some functions. Below, we actually call the functions. If you run this cell multiple times, you will get different maps each time, each of which have the same power spectrum. **This is important. When we're simulating a CMB, we're really simulating the power spectrum. *Not* the sky. Depending on the phase, two simulated skies will look completely different, even with the same power spectrum.**

Think of the CMB as similar to a patch of bark on two different trees of the same species. If we took images of the bark, we don't expect them to be *exactly* the same, we just expect them to have the same characteristics, with similarly sized shapes.

```python
## make a CMB T map
CMB_T = make_CMB_T_map(N,pix_size,axis,data_nominal)
p = Plot_CMB_Map(CMB_T,c_min,c_max,X_width,Y_width)
```
