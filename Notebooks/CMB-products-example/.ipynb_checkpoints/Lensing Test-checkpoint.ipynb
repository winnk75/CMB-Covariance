Dan and I are trying to figure out whether or not the discrepancy between the self-run CAMB power spectra and the theory $C_{\ell}$ power spectra provided by Planck is due to lensing, or something else entirely.

#### The Problem
The following three cells load data from a camb and isitgr cosmosis run and plot the log of the residuals. As you can see, the camb is only accurate to about 3 orders of magnitude (ie: one part in 1,000) and the isitgr is accurate to one part in 2 orders of magnitude.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Builds an x axis
# This creates an array 2.0, 3.0 ... 2508.0
x = np.arange(2.0,2509.0,1.0)
z = np.zeros(2507)

```

```python
#File paths to copies of previous cosmosis runs and planck provided run,.,..

camb_data = '/Users/winn/Research/Thesis/CMB-products/lensing-test-data/problem-camb.txt'
isitgr_data = '/Users/winn/Research/Thesis/CMB-products/lensing-test-data/problem-isitgr.txt'
planckdata = '/Users/winn/Research/Thesis/CMB-products/lensing-test-data/planck-power-spectra.txt'

a1 = np.loadtxt(camb_data,skiprows=1)
a2 = np.loadtxt(isitgr_data,skiprows=1)
pl = np.loadtxt(planckdata,skiprows=1,usecols=(1,))

f1 = a1[0:2507]
f2 = a2[0:2507]
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.log10(np.absolute((f1-pl)/pl)), color='r', ls='none', marker='.', markersize='1', label='Camb TT Spectrum')
ax.plot(x, np.log10(np.absolute((f2-pl)/pl)), color='b', ls='none', marker='.', markersize='1', label='ISiTGR TT Spectrum')
ax.plot(x, z, color='#26a192', label='Planck values_base_plikHM_tt_lowTEB')

ax.legend(loc='lower right')
ax.set_title("Log Residuals of CAMB and ISiTGR codes for Planck 2015 Theory Curve")
ax.set_xlabel('$\ell$')
```

```python
short = (f1[0:2000]-pl[0:2000])/pl[0:2000]

plt.plot(x[0:2000], np.log10(np.absolute(short)),label='Log Residuals CAMB')
plt.legend(loc='lower right')
plt.savefig("planck.pdf")
```

### The Test

Now that the problem is clear, we're going to test whether or not CMB lensing is the cause of this. While we've selected do_lensing=T for Camb, we're going to set it to false as well just to compare. I should also note that we're ignoring isitgr for now and just focusing on Camb.

The respective likelihoods provided by Cosomosis are -111 and -349 (lensed and unlensed), so already I can tell that they're different. Plot is below.

```python
# This cell loads the data and formats the arrays

file_camb_lensed = '/Users/winn/Research/Thesis/CMB-products/lensing-test-data/camb-lensed.txt'
file_camb_unlensed = '/Users/winn/Research/Thesis/CMB-products/lensing-test-data/camb-unlensed.txt'

data_camb_lensed = np.loadtxt(file_camb_lensed,skiprows=1)
data_camb_unlensed = np.loadtxt(file_camb_unlensed,skiprows=1)

data_camb_lensed = data_camb_lensed[0:2507]
data_camb_unlensed = data_camb_unlensed[0:2507]
```

```python
# Plot the data

plt.plot(x, data_camb_lensed,  color='r', ls='none', marker='.', markersize='1', label='Camb Lensed')
plt.plot(x, data_camb_unlensed,  color='b', ls='none', marker='.', markersize='1', label='Camb Unlensed')

plt.legend(loc='lower right')
```

As you can see these are almost the same, but they're not exactly. Plotted below is the numerical residuals plot.

```python
resid = data_camb_unlensed - data_camb_lensed
#print(resid)

plt.plot(x, data_camb_lensed-data_camb_lensed,  color='r', ls='none', marker='.', markersize='1')
plt.plot(x, resid,  color='b', ls='none', marker='.', markersize='1', label='Camb Unlensed Residuals')
plt.legend(loc='upper right')
plt.title('Residuals for Unlensed Power Spectrum')
plt.savefig('./unlensed-residuals.pdf', bbox_inches='tight')
```

```python
plt.plot(x, np.log10(np.absolute((resid)/data_camb_unlensed)), color='b', ls='none', marker='.', markersize='1', label='Unlensed Log Residuals')
plt.legend(loc='upper left')
plt.title('Log Residuals for Unlensed Power Spectrum')
plt.savefig('./unlensed-log-residuals.pdf', bbox_inches='tight')
```
