I mentioned that in this notebook we will be exploring "how well we know what we think we know." To do this, we will be using Fisher Analysis. In Fisher Analysis, we use *how much* some observable data set changes (in this case CMB power spectra) when we tweak some underlying parameter (in this case the MG parameter $D_0$), and compare this with the errors on the observable data set (observational uncertainty in the CMB). Ultimately, we can infer the the *best possible* constraint on the underlying $D_0$ parameter from the CMB power spectra uncertainty.

I want to point out that this isn't a constraint on $D_0$ *per se*, just the theoretical best possible constraint you could get through, say, an MCMC. This is a subtle but important point. So why is this helpful? Well, as you run better and better MCMC chains, you should expect to get closer and closer to this Fisher result, but not pass it. In that way, it's an analytic sanity check for computational methods.

### Basic Fisher Constraint

We're going to start with the basic "Fisher Constraint," which is pretty much a 1x1 Fisher Matrix (which is more general). The Fisher Constraint is exactly what I've described in the previous paragraph: a constraint on a parameter. The Fisher Matrix, on the other hand, is a multi dimensional covariance matrix. So this is more helpful if we wanted to look for, for example, degeneracies between $D_0$ and other cosmological parameters.

Anyway, let's introduce the Fisher Constraint with the following equation:

$F \ = \ \sum_k \big( \frac{\partial D_k}{\partial P} \big)^2 \ \frac{1}{\sigma_k^2} $

We sum across all data points, $k$, taking the partial derivative of the entire data set, $D_k$, with respect to a change in parameter, $P$ (in our case this will be $D_0$). $\sigma_k$ is the uncertainty in the data, which for us will be cosmic variance and instrument noise. Ultimately, the Fisher Constraint is defined as:

$\sigma_P = F^{-1/2}$

So let's break down each term. The sum across data points, $k$, will be each $\ell$ value for all three of the CMB power spectra we are using. So $D_k$ will be $D_{\ell}$ for the power spectra. The partial derivative, $\frac{\partial D_k}{\partial P}$ we will calculate from the difference between two power spectra (at slightly different $D_0$ values). The various power spectra we plotted in the previous notebook are not just for show; we are using these to calculate numerical derivatives.

For $\sigma_k$, the CMB uncertainty, we will use cosmic variance and instrument noise. Cosmic variance is easily computed using the following, for TT or EE (no cosmic variance for TE):

$\sigma_{\ell} \ = \ C_{\ell} \sqrt{ \frac{2}{2 \ell + 1} }$

Instrument noise is somewhat more complicated, but a rough approxomation may be used (below). This expression uses the FWHM of the gaussian beam, so the smaller the angle, the higher the noise. This is from arXiv: 1505.00639.


$N_{\ell}^{XZ} \ = \ \delta^2_{XZ} e^{\ell (\ell +1) \theta^2_{FWHM} / 8 \ln(2)}$

...where $\delta_{XZ} = 0$ for $X \neq Z$.

Now we'll add error. To make things easy, we just build two functions. The Noise and FWHM default values are for *Planck* at the 217 GHz channel. All values are available in the paper I mentioned. Finally, we plot both error contributions just to see. These errors are uncorrelated, so they may be freely added in quadrature. 

```python
# Let's actually define some functions here...
# The axis and data arguments take lists/arrays, and the function spits out a list of errors (for each ell)

def cosmic_variance(axis,data):
    return data*np.sqrt( 2. / (2.*axis + 1.) )

def instrument_noise(axis,data,noise=54.,fwhm=5.):
    # Default values are for Planck at 217 GHz; FWHM = 5.0 [arcmin]     NOISE = 54.0 [uK arcmin]
    
    # Conversion from uK arcmin into uK radian (for noise)
    # This gets squared below
    noise = noise*(np.pi/180.)*(1./60.)
    
    # Conversion from arcmin into radian (for beam fwhm)
    fwhm = fwhm*(np.pi/180.)*(1./60.)
    
    # need to convert from C_L to D_L so multiple the whole enchelada by (L(L+1)) / 2pi...
    return ((axis*(axis+1.))/(2.*np.pi))*(noise**2.)*np.exp( (axis*(axis+1.)*(fwhm**2.))/(5.545177) )

    # As a reminder, this all follows from 1505.00639 (see table 3 and eqn 47)

def combined_noise(axis,data,noise=54.,fwhm=5.):
    return np.sqrt( instrument_noise(axis,data,noise,fwhm)**2. + cosmic_variance(axis,data)**2. )

print(instrument_noise(axis,data_nominal))

plt.errorbar(axis, data_nominal, yerr=cosmic_variance(axis,data_nominal),label='Cosmic Variance',linewidth=1)
plt.errorbar(axis, data_nominal, yerr=instrument_noise(axis,data_nominal),label='Instrument Noise',color='#f48c3f',linewidth=1)
#plt.errorbar(axis, data_nominal, yerr=combined_noise(axis,data_nominal),label='Combined',linewidth=1)

#plt.errorbar(axis[0::15], data_nominal[0::15], yerr=combined_noise(axis,data_nominal)[0::15],label='Combined, every 15th bin',linewidth=1)

plt.xscale('log')
#plt.yscale('log')

#plt.xlim(0,1500)
#plt.ylim(0,6000)

plt.title('Projected Noise Contributions')
plt.xlabel('$\ell$')
plt.ylabel(r'$\frac {\ell (\ell+1)} {2 \pi}$ $C_\ell$  [$\mu$K$^2$]')
plt.legend(loc='upper left')
plt.ylim(-200.,6900.)
#plt.tight_layout()
#plt.savefig('combined_error_a.pdf')
```
